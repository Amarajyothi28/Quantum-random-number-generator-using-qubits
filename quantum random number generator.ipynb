```python
from qiskit import QuantumCircuit
from qiskit_aer import Aer
```

```python
qc= QuantumCircuit(1,1) #1 qubit,1 classical bit
qc.h(0) #apply Hadamard gate on qubit 0
qc.measure(0,0)#measure qubit 0->classical bit 0
```

```python
simulator = Aer.get_backend('qasm_simulator')
result = simulator.run(qc, shots=1).result()
counts = result.get_counts()
print("Random Number (0 or 1):",list(counts.keys())[0])
```

```python
!pip install qiskit-aer
from qiskit import QuantumCircuit
from qiskit_aer import Aer
import matplotlib.pyplot as plt

# Define the quantum circuit (previously in cell aWRgUibQqulW)
qc = QuantumCircuit(1, 1) # 1 qubit, 1 classical bit
qc.h(0) # apply Hadamard gate on qubit 0
qc.measure(0, 0) # measure qubit 0->classical bit 0

simulator = Aer.get_backend('qasm_simulator')
result = simulator.run(qc, shots=1).result()
counts = result.get_counts()

# Prepare data for bar graph
bits = list(counts.keys())
frequency = list(counts.values())

# Plot bar graph
plt.bar(bits, frequency)
plt.xlabel("Random Bit (0 or 1)")
plt.ylabel("Frequency")
plt.title("Quantum Random Number Generator - Bar Graph")
plt.show()
```
